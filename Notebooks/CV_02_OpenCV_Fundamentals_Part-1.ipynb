## Import required libraries

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

import matplotlib

from google.colab.patches import cv2_imshow
```

### Look at versions of various libraries

```
print('OpenCV: ', cv2.__version__)
print('Numpy: ', np.__version__)
print('Matplotlib: ', matplotlib.__version__)
```

## Read and Display the image

```
# imread(filename[, flags]) -> retval
#-    @brief Loads an image from a file.

img1 = cv2.imread(filename="imgOCV/forest.jpg", flags=1)

# flags=1 reads the image as color image
# flags=0 reads the image as gray-scale

# img1 is a numpy array
```

## Display the image

```
# Option 1 - Using Colab builtin function

print('Original image\n')
cv2_imshow(img1)
```

```
# Option 2 - Using Matplotlib

# NOTE: cv2 reads the image in BGR format, but typical format is RGB.

# When we want to display color images read using cv2 with plt,
# then the order of the color channels need to be reversed.

img1_RGB = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.imshow(img1_RGB)
plt.title('Forest color image')
plt.show()
```

```
# If the conversion was not done, then the image looks as follows:

plt.imshow(img1)
plt.show()
```

## To view the grayscale image

- When reading the image, we could have provided __flags=0__.
- If already opened as color image, then it can be converted to grayscale.

```
img1_Gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

print('Grayscale image')
cv2_imshow(img1_Gray)
```

### If you want to see two images side-by-side, we can make use of matplotlib

```
# Subplotting using matplotlib.

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_figheight(12)
fig.set_figwidth(12)

ax1.set_title('img1_RGB')
ax1.imshow(img1_RGB)

ax2.set_title('img1_Gray')
ax2.imshow(img1_Gray, cmap='gray')
```

## Display some properties

```
# Type and Shape of the matrix

print('Type of img1: ', type(img1))

print('\n')

print('Shape of img1: ', img1.shape)
print('Shape of img1_Gray: ', img1_Gray.shape)

print('\n')

print('Data type of img1: ', img1.dtype)
print('Data type of img1_Gray: ', img1_Gray.dtype)
```

```
# Storing the attributes in separate variables.

(height, width, depth) = img1.shape
# image.shape returns the following:
# height: number of rows
# width : number of columns
# depth : number of color channels
#         3 for color images (Blue, Green, Red)

print(" Height: %d\n Width: %d\n Depth: %d" % (height, width, depth))
```

```
# Storing the attributes in separate variables.

(height, width) = img1_Gray.shape
# image.shape returns the following:
# height: number of rows
# width : number of columns
# depth : number of color channels
#         3 for color images (Blue, Green, Red)

print(" Height: %d\n Width: %d" % (height, width))
```

### Accessing individual pixels
