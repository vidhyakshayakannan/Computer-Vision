# Color-space

## Typical color space used in openCV is the RGB color-space.
- We can convert them to other color spaces

```
# Get the data from VisionCog website

! unzip imgOCV.zip
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('imgOCV/forest.jpg')
cv2_imshow(img)
```

## Gray-scale

```
img_Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print('Gray scale version \n')
cv2_imshow(img_Gray)
```

## HSV representation

```
img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

print('HSV version')
cv2_imshow(img_HSV)
```

## HSV to BGR

```
img_HSV_to_BGR = cv2.cvtColor(img_HSV, cv2.COLOR_HSV2BGR)

print('BGR version from HSV')
cv2_imshow(img_HSV_to_BGR)
```

## EXERCISE:

- Split HSV into individual channels and visualize

```
# Convert img to img_HSV once again.
# Split it into H-Channel, S-Channel and V-Channel.
# Display all the channels together with original side-by-side.

img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

print(img_HSV.shape)
(H_Channel, S_Channel, V_Channel) = cv2.split(img_HSV)

print(H_Channel.shape)
```

```
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)

fig.set_figheight(20)
fig.set_figwidth(20)

ax1.set_title('Color Image')
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

ax2.set_title('H-Channel')
ax2.imshow(H_Channel, cmap='gray')

ax3.set_title('S-Channel')
ax3.imshow(S_Channel, cmap='gray')

ax4.set_title('V-Channel')
ax4.imshow(V_Channel, cmap='gray')
```

## L* a* b*

```
img_Lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

cv2_imshow(img_Lab)
```

## EXERCISE:

- Split L* a* b* into individual channels and visualize

```
# Split img_Lab into L_Channel, a_Channel and b_Channel.
# Display all the channels together with original side-by-side.

(L_Channel, a_Channel, b_Channel) = cv2.split(img_Lab)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)

fig.set_figheight(20)
fig.set_figwidth(20)

ax1.set_title('Color Image')
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

ax2.set_title('L-Channel')
ax2.imshow(L_Channel, cmap='gray')

ax3.set_title('a-Channel')
ax3.imshow(a_Channel, cmap='gray')

ax4.set_title('b-Channel')
ax4.imshow(b_Channel, cmap='gray')
```
